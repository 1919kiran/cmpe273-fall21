```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load
import random
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
from tensorflow import keras

#df = pd.read_csv('../input/regressionhousing/Housing.csv')
#df_copy = df.copy(deep=True)
#df.head()

features = np.array([1, 2, 3, 5, 6, 7], dtype=float)
labels = np.array([155, 197, 244, 356, 407, 448], dtype=float)

def square_trick(base_price, weight, num_rooms, price, learning_rate):
    predicted_price = base_price + weight * num_rooms
    base_price += learning_rate * (price - predicted_price)
    weight += learning_rate * num_rooms * (price - predicted_price)
    return weight, base_price
    
def linear_regression(features, labels, learning_rate=0.01, epochs=10):
    weight = random.random()
    base_price = random.random()
    for epoch in range(epochs):
        i = random.randint(0, len(features)-1)
        num_rooms = features[i]
        price = labels[i]
        weight, base_price = square_trick(base_price, weight, num_rooms, price, learning_rate)
    return weight, base_price


def predict_house_price(num_rooms, epochs):
    weight, base_price = linear_regression(features, labels, learning_rate=0.01, epochs=epochs)
    print(f"weight={weight}, base_price={base_price}")
    predicted_price = weight * num_rooms + base_price
    print(f"Predicted pirce for house with {num_rooms} rooms:{round(predicted_price, 2)}")

    
def predict_naive(num_rooms, epochs):
    weight, base_price = linear_regression(features, labels, learning_rate=0.01, epochs=epochs)
    print(f"weight={weight}, base_={base_price}")
    predicted_price = weight * num_rooms + base_price
    print(f"[Naive]Predicted value: {round(predicted_price, 4)}")
    
## Kera Tensorflow Linear Regression
def keras_linear_regression():
    model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(features, labels, epochs=500)
    return model

def keras_predict(model, x):
    print(f"[TF]Predicted value: {model.predict(x)}") #[10.0]

def rmse(labels, predictions):
    n = len(labels)
    differences = np.subtract(labels, predictions)
    return np.sqrt(1.0/n * np.dot(differences, differences))
    
# Keras TF Linear Regression
def tf_predictions():
    tf_model = keras_linear_regression()
    predictions = []
    for n in labels:
        predictions.append(keras_predict(tf_model, [n]))
    
    return predictions
print(f"TF_predictions:{tf_predictions}")

# Naive Linear Regression
def naive_predictions():
    predictions = []
    for n in labels:
        predictions.append(predict_naive(n, 500))
        
    return predictions

print(f"Naive_predictions:{naive_predictions()}")

# TODO: RMSE with test dataset
```
